# tikz-network

Remember that we have to use the package `tikz-network`. This can be done by passing the `-p tikz-network` option to the `%%tikz` cell magic

## First Graph - Vertex Locations

```python
%%tikz -s 200,200 -f svg -p tikz-network
\Vertex{A}
\Vertex[x=1,y=1]{B}
\Vertex[x=2]{C}
```

## Vertex Labels

```python
%%tikz -s 200,200 -f svg -p tikz-network
\Vertex[label=$x_1$]{A}
\Vertex[label=$x_2$,x=1,y=1]{B}
\Vertex[label=$x_3$,x=2]{C}
```

## Pseudo Vertices

```python
%%tikz -s 200,200 -f svg -p tikz-network
\Vertex[label=$x_1$]{A}
\Vertex[label=$x_2$,x=1,Pseudo]{B}
\Vertex[label=$x_3$,x=2]{C}
```

## Tensor With Four Legs

```python
%%tikz -s 200,200 -f svg -p tikz-network
\Vertex[label=R]{R}
\Vertex[y=1,Pseudo,label=$\rho$]{rho}
\Vertex[x=-1,y=-1,Pseudo,label=$\sigma$]{sigma}
\Vertex[y=-1,Pseudo,label=$\mu$]{mu}
\Vertex[x=1,y=-1,Pseudo,label=$\nu$]{nu}
\Edge(R)(rho)
\Edge(R)(sigma)
\Edge(R)(mu)
\Edge(R)(nu)
```

# Single Vertex

```python
%%tikz -s 200,200 -f svg
\tikzset{vertex_style/.style={fill=blue!50,circle}}
\node [vertex_style] (R) {R};
```

# Relative Positioning using `positioning` library

```python
%%tikz -s 200,200 -f svg -l positioning
\tikzset{vertex_node/.style={fill=blue!50,circle}}
%\tikzset{hidden_node}/.style={fill=white}
\node [vertex_node] (R) {R};
\node (rho) [above=of R] {$\rho$};
\node (sigma) [below left=of R] {$\sigma$};
\node (mu) [below=of R] {$\mu$};
\node (nu) [below right=of R] {$\nu$};
```

```python
%%tikz -l graphs,positioning -s 200,200 -f svg 
\tikzset{vertex_node/.style={fill=blue!50,circle}}
%\tikzset{hidden_node}/.style={fill=white}
\node [vertex_node] (R) {R};
\node (rho) [above=of R] {$\rho$};
\node (sigma) [below left=of R] {$\sigma$};
\node (mu) [below=of R] {$\mu$};
\node (nu) [below right=of R] {$\nu$};
\graph [use existing nodes]{
  R -- rho;  
};
```

## Important Note

`tikz-magic` is very particular about spacing. `-l graphs, positioning` gives an error, but `-l graphs,positioning` works fine.

# Tensor With Four Legs

Corresponds to eqn 1.1 in Bridgeman.

```python
%%tikz -l graphs,positioning -s 200,200 -f svg 
\tikzset{vertex_node/.style={fill=blue!50,circle,scale=2.0}}
%\tikzset{hidden_node}/.style={fill=white}
\node [vertex_node] (R) {R};
\node (rho) [above=of R] {$\rho$};
\node (sigma) [below left=of R] {$\sigma$};
\node (mu) [below=of R] {$\mu$};
\node (nu) [below right=of R] {$\nu$};
\graph [use existing nodes]{
    R -- rho;
    R -- sigma;
    R -- mu;
    R -- nu;
};
```

## Two Tensors with Free Legs

```python
%%tikz -l graphs,positioning -s 200,200 -f svg 
\tikzset{vertex_node/.style={fill=blue!50,circle}}
%\tikzset{hidden_node}/.style={fill=white}
\tikzset{help lines/.style={color=blue!50,very thin}}
\tikzset{node distance=1cm}
\draw [help lines, step=1cm] (-2,-2) grid (2,2);
\node [vertex_node] (A) {A};
\node at (2,0) [vertex_node] (B) {B};
\path (A) edge[-] (-1,1);
\path (A) edge[-] (1,1);
\path (A) edge[-] (0,-1);
```

```python
%%tikz -l graphs,positioning -s 200,200 -f svg 
\tikzset{vertex_node/.style={fill=blue!50,circle}}
%\tikzset{hidden_node}/.style={fill=white}
\tikzset{help lines/.style={color=blue!50,very thin}}
\tikzset{node distance=1cm}
\draw [help lines, step=1cm] (-2,-2) grid (2,2);
\node [vertex_node] (A) {A};
\node at (2,0) [vertex_node] (B) {B};
\path (A) edge[-] (-1,1);
\path (A) edge[-] (1,1);
\path (A) edge[-] (0,-1);
\path (B) edge [-] (2,1);
\path (B) edge [-] (2,-1);
```
